```python
# 집합
# 집합은 {}를 사용해서 선언하는 자료형으로 중복을 허용하지 않음.
# 원소의 순서가 의미없으므로 인덱싱/슬라이싱 불가
# 순서가 의미 없기 때문에 리스트는 원소로 못 넣는다.

# {} 사용해서 집합 만들기
name = {'이정재', '박해수', '오영수', '위하준', '정호연', '허성태', '미주령'}
print(type(name))

# 공집합을 만들 때는 {}로 하면 딕셔너리가 되므로 안 되고, set을 이용한다.
name1 = {}
print(type(name1))

# 공집합
name2 = set()
print(type(name2))
```

```python
# [1,2]라는 리스트는 순서 0번째, 1번째가 있기 때문에 set이 될 수 없다.
names = {'이정재', '박해수', '정호연', [1,2]}
print(len(names))
print(names)
```

```python
# 집합에서 원소 추가하기 
# .add(): 하나의 원소 추가
names = {'이정재', '박해수', '오영수', '위하준', '정호연', '허성태', '미주령'}
names.add('압둘 알리')
print(names) # 순서 중요하지 않다.

# .update(): 여러 개의 원소 추가
# 리스트, 튜플, 딕셔너리를 넣을 수 있는데, 다 원소로 분해되어 들어감.
names.update(('고길동', '둘리'))
names.update(['표창원', '허성태'])
names.update({1,2,3})

print(names) # 마찬가지로 순서 중요하지 않다.
```

```python
# 집합에서 원소 제거하기
# 1. remove(): 집합에 없는 원소를 지우려 하면 에러 발생
names = {'이정재', '박해수', '오영수', '위하준', '정호연', '허성태', '미주령'}
names.remove('이정재')
print(names)

# 2. discard(): 집합에 없는 원소를 지우려 해도 에러 발생하지 않음
names.discard('박해수')
print(names)

# 그렇다면 무조건 remove가 아닌 discard를 쓰는 것이 좋을까?
# 만약 오류가 발생한다면, 다른 플랜을 짤 수 있도록 하기 위해 remove가 필요할 수도 있다.
# 따라서 둘 다 알고 있어야 한다.

# 3. pop(): 랜덤하게 요소 제거 후 반환
print(names.pop()) # 랜덤 요소 출력
print(names)

# 4. clear(): 공집합으로 만듦
names.clear()
print(names)
```

```python
# 집합의 연산
# 합집합: |(or) 또는 union 메소드
# 교집합: &(and) 또는 intersection 메소드
# 차집합: - 또는 difference 메소드 
# 대칭 차집합: ^ 또는 symmetric_difference 메소드 사용
a = {1,2,3,4}
print(type(a))
b = set([3,4,5,6])
print(type(b))

# 합집합
print(a|b)
print(a.union(b))

# 교집합
print(a&b)
print(a.intersection(b))

# 차집합
print(a-b)
print(b-a)
print(a.difference(b))

# 대칭차집합
print(a^b)
print(a.symmetric_difference(b))
```

```python
# 인덱스의 위치도 알려주는 enumerate()
name = ['이정재', '박해수', '오영수', '위하준', '정호연', '허성태']
for i, j in enumerate(name):
    print(i,j)
```

```python
# 리스트에서는 리스트 속 데이터에서 조건에 맞는 데이터를 가져오는 다른 방법이 있다.
# 약속된 구문으로 숙지해두자.
lst = list(range(10))
lst_odd = [x for x in lst if x%2==1]
lst_even = [x for x in lst if x%2==0]

print(lst_odd)
print(lst_even)
```

```python
# 아래와 같이 for문과 if문을 사용할 수도 있지만 위에 구문이 편리하다.
lst = list(range(10))
lst_odd = []
lst_even = []
for i in range(10):
    if lst[i]%2 == 0:
        lst_even.append(lst[i])
    else:
        lst_odd.append(lst[i])

print(lst_odd)
print(lst_even)
```

## 함수

```python
def sum1(a,b):
    return a+b

sum1(1,2)
```

```python
# 여러 개의 값을 한번에 받을 때, 함수가 반환하는 값이 여러 개일 경우 개수에 맞춰 변수를 할당해줘야 함.
# 원하지 않는 값일 경우 '_' 표시로 반환을 생략할 수 있음
# 하나의 변수로 여러 값을 받을 때는 하나의 변수가 여러 값을 지닌 튜플이 됨 => 변하면 안 되니까!
def sum1(a,b):
    return a,b,a+b

# 1. 반환값 개수에 맞춰 변수 개수 입력한 경우
d,e,f = sum1(1,2)
print(d,e,f)

# 2. 반환값은 3개인데 변수 1개일 경우 
g = sum1(1,2)
print(g) # 튜플
print(type(g))
```

```python
def sum2(a,b):
    if type(a) == int and type(b) == int:
        return a+b
    else:
        print('숫자를 입력하세요.')

print(sum2(2,3))
sum2('r',3)
```

```python
# 함수로 보내는 파라미터 개수가 정해져 있어야 할까? (가변 매개변수)
# 입력하는 변수가 10개일 때, 10개를 전부 함수 input으로 적어주는 것이 아닌 *를 사용한다. 
# args라고 꼭 안 써줘도 된다. *이 중요함.
def sum2(*args): # 튜플로 들어간다.
    return sum(args), type(args)

sum2(1,2,3,4,5,6) # 튜플로 받는다.
```

```python
# 함수를 만들 때 하나의 변수+튜플을 input으로 받으려면 반드시 순서를 지켜야 한다.
# 함수 호출 시 첫 번째 인자는 반드시 name으로 들어간다. 나머지는 전부 args에 튜플 형태로 저장된다.
def exer(name, *args):
    print(name)
    print(len(args))
    print(type(args))

exer('이상훈',1,2,3)
exer(1,2,3,4,5,'이상훈')
```

```python
# 실행이 안 됨.
# 파이썬 함수 설계될 때, 법칙.
def exer(*args, name): # 변수 다음 튜플 순서로 해야함
    print(name)
    print(len(args))
    print(type(args))


exer('이상훈',1,2,3)

# 만약 딕셔너리까지 넣고 싶다면 (변수, 튜플, 딕셔너리) 순서로.
```

## 클래스
- 클래스를 사용하는 이유? 비슷한 데이터와 기능을 하나의 틀로 묶어서, 깔끔하고 재사용 가능하게 코드를 관리하기 위해
- **클래스의 상속**
    - 상속은 보통 기존 클래스를 변경하지 않고 기능을 추가하거나 기존 기능을 변경할 때 사용
    - 기존 클래스가 라이브러리 형태로 제공되거나 수정이 허용되지 않는 상황이라면 상속을 해야함

- \__init__\: 클래스 시작할 때 꼭 받아야 하는 변수

```python
class coin:
    def __init__(self,name):
        self.name = name
        self.shares = 0
    
    def buy(self, shares):
        self.shares+=shares
        print(self.name + " 코인 {} 개 매수 완료".format(shares))

    def sell(self, shares):
        self.shares-=shares
        print(self.name + ' 코인 {} 개 매도 완료'.format(shares))

    def balance(self):
        print('현재 '+self.name+' 코인은 {} 개 보유 중입니다.'.format(self.shares)) # 계속 계산되고 있는 self.shares 임
```

```python
# 클래스 상속
class add_coin(coin):
    def sell_all(self):
        self.shares = 0
        print(self.name + ' 코인 전량 매도 완료')
```

```python
eos = add_coin('EOS')
eos.buy(3)
eos.sell_all()
eos.balance()
```

```python
Dog = coin('Dog')
Dog.buy(10)
Dog.sell(3)
```

```python
# *args는 튜플로 받아지고, 튜플은 수정이 불가능하기 때문에 수정이 안 된다? 
# 리스트로 바꾸면 된다.
class squid_game:
    def __init__(self, *args): # 튜플로 받아온다
        self.data = list(args)[0] # 리스트로 만든다
        print('참가자 명단은 아래와 같습니다. \n'+str(self.data))

    def check(self):
        print(self.data)
        print(type(self.data))
        return self.data
    
    def fail(self, *args):
        self.data = [x for x in self.data if x not in list(args)[0]] # 튜플을 리스트로 바꿔줌
        print(args)
        print(type(self.data))
        return self.data
    
    def fail2(self, *args): # 위에 fail 함수랑 똑같은데, 다른 코드 사용
        args = list(args)[0]
        print(len(args))
        for i in range(0,len(args)):
            print(args[i])
            self.data.remove(args[i])
        return self.data

    def night_random_fight(self, num):
        for i in range(num):
            j = self.data.pop()
            print('{} 번째 참가자 사망'.format(j))
        return self.data
```

```python
# 참가자
first_list = list(range(1,51))
first_comp = squid_game(first_list)
```

```python
# Try & Except
result = [8,9,10,'',11,12,'',23]

for i in result:
    try:
        if i>10:
            print(i)
    except:
        print('대소비교불가')
```

```python
for i in result:
    try:
        if i>10:
            if i==11:
                print('11 발견')
                pass # break는? continue는?
            print(i) # 핵심
    except:
        print('대소비교불가')
        continue

# pass 자리에 continue를 넣으면 이 다음 코드는 실행 안 되고 반복문으로 돌아간다.
# pass일 경우, 그 다음 코드로 그냥 넘어감. 
# break는 바로 멈춘다.
# pass: 그대로 진행해 / continue: 멈추고 다시 다음 루프로 돌아가라
```

```python
for i in result:
    try:
        if i>10:
            if i==11:
                print('11발견')
                raise Exception('에러') # 강제로 예외 처리하여 print(i) 결과 안 나오고 다음 단계로 넘어가도록 함.
            print(i)
    except Exception as e: # 파이썬에서 어떤 에러가 발생하고 있는지 그대로 출력하라
        print(e)
        # print('대소비교불가')
```

```python
z = np.random.randint(0,100,size=(10,10)) # 10x10
p = np.random.randint(0,100,size=(10,5)) # 10x5

print(np.multiply(z,p)) # 불가능
print(np.dot(z,p)) # 가능

# np.dot(z,p)는 가능하고 np.multiply(z,p)는 안 되는 이유: Broadcasting 이라는 개념
# Broadcasting: 차원이 달라도 1이거나 같으면 자동으로 늘려서 연산해주는 넘파이 규칙 (서로 다른 형태의 배열을 자동으로 늘려서 연산 가능하게 만듦)
# 오른쪽 차원부터 비교하면서, 값이 같거나 둘 중 하나가 1이어야 한다.
# 브로드캐스팅은 반복문 없이 벡터화 연산을 가능하게 해서, 연산 속도를 극적으로 높이고 메모리 사용을 줄이며, 수학적 연산을 코드로 자연스럽게 표현하기 위해 필요한 핵심 개념이다.
```
