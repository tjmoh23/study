## Lambda Function
- 저장 공간을 쓰지 않고 함수를 사용할 수 있다. (간단하게 함수를 사용할 수 있음)

```python
# 기본 사용 방식
func = lambda x,y : x+y
func(3,5)
```

```python
# 함수를 반환하는 함수를 정의한다고 했을 때
def out_func(in_func, num1, num2):
    return in_func(num1, num2)

# plus 함수를 정의해서 이렇게 사용할 수도 있고
def plus(a,b):
    return a+b

print(out_func(plus, 3, 5))

# lambda 함수 사용해서 바로 사용할 수도 있다.
out_func(lambda x,y:x+y, 5, 10)
```

## map / filter / reduce
### map
- 하나의 함수를 input으로 받은 후, 순서가 있는 집합 활용
- map(func, *iterable)
- func(iterable[0]), func(iterable[1]), func(iterable[2]), ... func(iterable[-1])
- zip(iterable)
- ls1 = [1,2,3] ls2 = [4,5,6]
- list(zip(ls1, ls2)) = [(1, 4), (2, 5), (3, 6)]

```python
ls1 = [1,2,3]
ls2 = [4,5,6]

list(map(lambda *args: args[0]+args[1], ls1, ls2))
```

```python
# lambda num : '홀수' if num % 2 else '짝수' 이게 하나의 함수임.
list(map(lambda num : '홀수' if num % 2 else '짝수', ls2))
```

```python
func = lambda num : '홀수' if num % 2 else '짝수'
[func(ls2[0]), func(ls2[1]), func(ls2[2])]
```

```python
[func(ls2[i]) for i in range(len(ls2))]
```

### filter
- filter(func, iterable)
    - func의 반환값이 반드시 boolean 형태여야 한다 (True/False)

```python
iterable = [1,2,3,4,5]
         # [True, False, True, False, True]
list(filter(lambda x : True if x % 2 else False, iterable))
```

```python
# 이렇게 def 함수 선언해서 출력해도 값 똑같다.
def judge(num):
    return True if num % 2 else False

list(filter(judge, iterable))
```

### reduce

```python
from functools import reduce
```

# 클래스
1. 클래스 선언
2. 상속 / 메소드 오버라이딩 / super 함수

```python
class FreindlySoldier:
    def __init__(self, hp, ad): # 생성자 함수
        self.hp = hp
        self.ad = ad

    def attack(self, enemy):
        enemy.hp -= self.ad


class EnemySoldier:
    def __init__(self, hp, ad): # 생성자 함수
        self.hp = hp
        self.ad = ad

    def attack(self, enemy):
        enemy.hp -= self.ad

# 생성자 함수 __init__: 클래스 생성할 때 자동으로 만들어짐
# 여기서의 self = FreindlySoldier를 의미. 자기 자신. 따라서 클래스 부르 거나 아래 함수 사용할 때 자기자신 넣어줄 필요 없음.
# 클래스 불러 올 때, FreindlySoldier(hp, ad)만 넣으면 된다. init에 맞에 넣어줘야 함.
# A.attack(B) => B.hp - A.ad 가 되는 거임.
```

```python
A = FreindlySoldier(100, 10)
B = EnemySoldier(30, 20)
```

```python
print(A.hp, A.ad)
print(B.hp, B.ad)
```

```python
# attack 안에 출력 반환하는 값 없기 때문에 결괏값 안 나온다.
A.attack(B)
```

```python
print(A.hp, A.ad)
print(B.hp, B.ad)
```

## 상속
- 자식 클래스는 부모 클래스의 기능을 이어받는다.

```python
class ComputingTool:
    def __init__(self, num1, num2):
        self.num1 = num1
        self.num2 = num2

    def add(self): # 이미 생성할 때 숫자 받기 때문에 굳이 또 num1, num2 받을 필요가 없다.
        return self.num1 + self.num2
```

```python
tool = ComputingTool(20, 50)
tool
```

```python
tool.num1, tool.num2, tool.add()
```

```python
# 상속
class ComputingTool2(ComputingTool):
    def sub(self):
        return self.num1 - self.num2
```

```python
tool2 = ComputingTool2(50,20)
tool2.sub(), tool2.add()
```

```python
# 메소드 오버라이딩
# 상속받은 것 중에 수정을 하고 싶다면?
class ComputingTool3(ComputingTool2):
    def add(self):
        return self.num1 ** 2 + self.num2 ** 2
    
# 기존에 정의되어 있던 함수명 그대로 가져와서 수정하면 된다.
```

```python
tool3 = ComputingTool3(50,20)
tool3.add(), tool3.sub()
```

### 1. timestamp
- pd.Timestamp
- pd.to_datetime
- timestamp 보다는 리스트로 여러 인자 넣을 수 있는 to_datetime을 많이 사용한다.

```python
pd.Timestamp('2025-11-27')
pd.Timestamp(2025,11,27)

date = pd.Timestamp(2025,10,7)
print(date.year)
print(date.month)
print(date.weekday())
date.today()
```

```python
# pd.to_datetime
# 문자열 포맷으로 넣어줘야 한다.
# 여러 문자열 넣을 수 있다 => DatetimeIndex가 된다.
date = pd.to_datetime(['2025-10-07', '2025-11-27']) 
print(date)
print(date[0])
print(date[1])
print(date[0].year)
print(date[1].day)
```

```python
# pd.date_range
# 일자 범위로 출력 가능
print('월말: ', pd.date_range('2025-10',' 2026-5', freq='M')) 
print('월초: ', pd.date_range('2025-10',' 2026-5', freq='MS')) 
```
